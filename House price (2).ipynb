```python
from sklearn.preprocessing import PolynomialFeatures 
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
Input=[('scale',StandardScaler()), ('polynomial', PolynomialFeatures(include_bias=False)), ('model',LinearRegression())]
pipe=Pipeline(Input)
pipe.fit(z,y)
ypipe=pipe.predict(z)
from sklearn.metrics import r2_score
r_squared = r2_score(ypipe,y)
print('The R-square value is: ', r_squared)
```

```python
#using polynomial regression, improved R^2 but only fractionally
```

```python
#diving the data into training and test data.

y_data = y
x_data = z

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.10, random_state=1)

print("number of test samples :", x_test.shape[0])
print("number of training samples:",x_train.shape[0])


#polynomial transfromation of z
pr=PolynomialFeatures(degree=2)
x_train_pr=pr.fit_transform(x_train)
x_test_pr=pr.fit_transform(x_test)

#linear regression model to train it
poly = LinearRegression()
poly.fit(x_train_pr, y_train)

#predicting for test data
yhat = poly.predict(x_test_pr)
yhat[0:5]

print("Predicted values:", yhat[0:4])
print("True values:", y_test[0:4].values)

#R^2 for training set
rtrain = poly.score(x_train_pr, y_train)

#R^2 for test set
rtest = poly.score(x_test_pr, y_test)

print("\n\nR^2 for training data: ",rtrain, "\nR^2 for test data: ",rtest)


#trying to eliminate the effect of multicollinearity in this data using Ridge regression
#so that predicted price is much near actal price
from sklearn.linear_model import Ridge
RigeModel = Ridge(alpha=0.1)

RigeModel.fit(x_train_pr, y_train)

YP = RigeModel.predict(x_test_pr)
print("\n\narray: ",YP[0:4])
print('test set :', y_test[0:4].values)

rRigeTest = RigeModel.score(x_test_pr, y_test)

rRigeTrain = RigeModel.score(x_train_pr, y_train)
print("\n\nR^2 for training data: ",rRigeTest, "\nR^2 for test data: ",rRigeTrain)
```

```python
#however, in attempt to eliminate multicoliearity of attributs, the R^2 value suffered a decline.
```
